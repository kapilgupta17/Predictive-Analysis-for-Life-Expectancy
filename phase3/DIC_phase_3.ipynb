```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,precision_score,recall_score,auc,roc_curve
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import label_binarize
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import pickle
```

```python
main_df=pd.read_csv('Life Expectancy Dataset(cleaned_encoded_normalized).csv')
main_df = pd.DataFrame(main_df)
```

```python
main_df
```

```python
main_df.columns
```

```python
features=['Region','IncomeGroup','Prevelance of Undernourishment','Health Expenditure %','Unemployment','Income per capita','Adult Mortality(per 1000 people)','Education expenditure(%)','Population','GDP','Infant deaths(per 1000 live births)']
X=main_df[features]
Y=main_df['Life Expectancy World Bank']
```

```python
#Splitting the data
dfx_train,dfx_test,dfy_train,dfy_test = train_test_split(X,Y,test_size=0.2)
```

# Regression

# Random forest

```python
random_forest = RandomForestRegressor(n_estimators=100, random_state=42)
random_forest.fit(dfx_train,dfy_train)
rf_pred = random_forest.predict(dfx_test)
print("Mean Squared Error:",mean_squared_error(dfy_test, rf_pred))
print("R-squared (R^2):",r2_score(dfy_test, rf_pred))
```

```python
rf_filename = 'random_forest_model.pkl'
with open(rf_filename, 'wb') as model_file:
    pickle.dump(random_forest, model_file)
```

# Logistic Regression

```python
median_life_expectancy = dfy_train.median()
y_train_binary = (dfy_train > median_life_expectancy).astype(int)
y_test_binary = (dfy_test > median_life_expectancy).astype(int)
log_reg = LogisticRegression()
log_reg.fit(dfx_train, y_train_binary)
y_pred_binary = log_reg.predict(dfx_test)

accuracy_binary = accuracy_score(y_test_binary, y_pred_binary)
print(f"Accuracy: {accuracy_binary}")
print("Classification Report:")
print(classification_report(y_test_binary, y_pred_binary))
```

```python
lgr_filename = 'lgr.pkl'
with open(lgr_filename, 'wb') as model_file:
    pickle.dump(log_reg, model_file)
```

# Support Vector Regression

```python
svr = SVR(kernel='rbf', C=1.0, epsilon=0.06)
svr.fit(dfx_train,dfy_train)
svr_pred=svr.predict(dfx_test)
print("Mean Squared Error:",mean_squared_error(dfy_test, svr_pred))
print("R-squared (R^2):",r2_score(dfy_test, svr_pred))
```

```python
svr_filename = 'svr.pkl'
with open(svr_filename, 'wb') as model_file:
    pickle.dump(svr, model_file)
```

# Polynomial Regression

```python
degree = 2
poly = PolynomialFeatures(degree=degree)
X_train_poly = poly.fit_transform(dfx_train)
X_test_poly = poly.transform(dfx_test)

lin_reg = LinearRegression()
lin_reg.fit(X_train_poly, dfy_train)
lr_pred = lin_reg.predict(X_test_poly)

print("Mean Squared Error:",mean_squared_error(dfy_test, lr_pred))
print("R-squared (R^2):",r2_score(dfy_test, lr_pred))
```

```python
with open('poly_reg_model.pkl', 'wb') as model_file:
    pickle.dump({'poly': poly, 'regression_model': lin_reg}, model_file)
```

# Gradient Boosting Regression

```python
gbr_model = GradientBoostingRegressor(n_estimators=100,learning_rate=0.01, max_depth=5,min_samples_split=3,min_samples_leaf=3,random_state=41)
gbr_model.fit(dfx_train, dfy_train)
gbr_pred = gbr_model.predict(dfx_test)
print("Mean Squared Error:",mean_squared_error(dfy_test, gbr_pred))
print("R-squared (R^2):",r2_score(dfy_test, gbr_pred))
```

```python
gbr_filename = 'gbr.pkl'
with open(gbr_filename, 'wb') as model_file:
    pickle.dump(gbr_model, model_file)
```

# Classifiers

```python
features2=['Region','Life Expectancy World Bank','Prevelance of Undernourishment','Health Expenditure %','Unemployment','Income per capita','Adult Mortality(per 1000 people)','Education expenditure(%)','Population','GDP','Infant deaths(per 1000 live births)']
X2=main_df[features2]
Y2=main_df['IncomeGroup']
```

```python
#Splitting the data
dfx2_train,dfx2_test,dfy2_train,dfy2_test = train_test_split(X2,Y2,test_size=0.2)
```

# Logistic regression

```python
logreg = LogisticRegression(max_iter=1000, random_state=42)
logreg.fit(dfx2_train, dfy2_train)
log_pred = logreg.predict(dfx2_test)

accuracy_binary = accuracy_score(dfy2_test, log_pred)
print(f"Accuracy: {accuracy_binary}")
print("Classification Report:")
print(classification_report(dfy2_test, log_pred))
```

```python
lgr_filename = 'logreg.pkl'
with open(lgr_filename, 'wb') as model_file:
    pickle.dump(logreg, model_file)
```

# K-Nearest Neighbors

```python
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(dfx2_train,dfy2_train)
knn_pred=knn.predict(dfx2_test)
print('Accuracy:',accuracy_score(dfy2_test,knn_pred))
print('Classification Report:\n',classification_report(dfy2_test,knn_pred))
```

```python
knn_filename = 'knn.pkl'
with open(knn_filename, 'wb') as model_file:
    pickle.dump(knn, model_file)
```
